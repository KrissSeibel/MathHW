```python
x = []
y = []
for i in range(-100, 101):
    x_ = i/100
    x.append(x_)
    y.append(math.sqrt(abs(x_**2-1)))
plt.plot(x, y)
x1 = []
y1 = []
for i in range(-100, 101):
    x_ = i/100
    x1.append(x_)
    y1.append(-1*(math.sqrt(abs(x_**2-1))))
plt.plot(x1, y1)
```

```python
x = []
y = []
for i in range(-200, 201):
    x_ = i/100
    x.append(x_)
    y.append(math.sqrt(abs(1-(x_**2)/4)))
plt.plot(x, y)
x1 = []
y1 = []
for i in range(-200, 201):
    x_ = i/100
    x1.append(x_)
    y1.append(-1*math.sqrt(abs(1-(x_**2)/4)))
plt.plot(x1, y1)
```

```python
x = []
y = []
for i in range(-500, 501):
    x_ = i/100
    if x_<=-2 or x_>=2:
        x.append(x_)
        y.append(math.sqrt(abs((x_**2)-1)))
plt.plot(x, y)
x1 = []
y1 = []
for i in range(-500, 501):
    x_ = i/100
    if x_<=-2 or x_>=2:
        x1.append(x_)
        y1.append(-1*math.sqrt(abs((x_**2)-1)))
plt.plot(x1, y1)
```
